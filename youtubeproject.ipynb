```python
polarity=[]
for comment in USComments['textOriginal']:
    try:
       polarity.append(TextBlob(comment).sentiment.polarity)
    except:
        polarity.append(0)
```

```python
len(polarity)
```

```python
USComments['polarity'] = polarity
```

```python
USComments.head(5)
```

```python
filter1 = USComments['polarity']==1
```

```python
comments_positive=USComments[filter1]
comments_positive
```

```python
filter2 = USComments['polarity']==-1
```

```python
comments_negitive=USComments[filter2]
comments_negitive
```
